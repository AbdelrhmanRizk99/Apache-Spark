```python
import findspark
findspark.init()
```

```python
from pyspark import SparkContext, SparkConf
```

```python
conf = SparkConf().setMaster("local").setAppName("WordCount")
sc = SparkContext(conf=conf)
```

```python
inputFile = sc.textFile("C:/Users/Abdelrhman/OneDrive/Desktop/python/Sources/textFile.txt")
```

```python
words = inputFile.flatMap(lambda line : line.split(" "))

```

```python
words.take(3)
```

```python
counts = words.map(lambda word :(word,1))
```

```python
counts.collect()
```

```python
result = counts.reduceByKey(lambda a,b : a+b)
```

```python
result.collect()
```
